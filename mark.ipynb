```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Import seaborn for violin plots (requires installation)

# Load the dataset from the CSV file
df = pd.read_csv('/content/mark.csv')

# Create box plot for grades
plt.figure(figsize=(8, 6))
sns.boxplot(x='Grade', y='Test Score', data=df, order=['A', 'B', 'C', 'D', 'F'])
plt.title('Box Plot of Test Scores by Grade')
plt.xlabel('Grade')
plt.ylabel('Test Score')
plt.show()

# Create violin plot for grades
plt.figure(figsize=(8, 6))
sns.violinplot(x='Grade', y='Test Score', data=df, order=['A', 'B', 'C', 'D', 'F'])
plt.title('Violin Plot of Test Scores by Grade')
plt.xlabel('Grade')
plt.ylabel('Test Score')
plt.show()
```
